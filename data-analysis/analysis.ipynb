### PAWS

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df_labeled_paws = pd.read_csv("../datasets/labeled/google/google_labeled_paws.csv")
df_labeled_paws.head()
```

```python
df_labeled_paws[df_labeled_paws['sentence_2_form'] == 3]
```

```python
def generate_graph(x, title):
    g = sns.catplot(kind="count", data=df_labeled_paws, x=x, palette="Set2", hue=x, legend=False)
    g.set_xticklabels(["Di Karaniwang Ayos", "Karaniwang Ayos"]) 
    g.set_axis_labels("Sentence Form", "Count")
    g.figure.suptitle(title)
    g.set(yticks=np.arange(0, df_labeled_paws[x].value_counts().max(), 20))
    plt.grid()
    plt.show()
```

```python
generate_graph("sentence_1_form", "PAWS: Sentence 1 Forms")
```

```python
g = sns.catplot(kind="count", data=df_labeled_paws, x="sentence_2_form", palette="Set2", hue="sentence_2_form", legend=False)
g.set_xticklabels(["Di Karaniwang Ayos", "Karaniwang Ayos", "Ambiguous"]) 
g.set_axis_labels("Sentence Form", "Count")
g.figure.suptitle("PAWS: Sentence 2 Form")
g.set(yticks=np.arange(0, df_labeled_paws["sentence_2_form"].value_counts().max(), 20))
plt.grid()
plt.show()
```
