# Cleaning Balanced COPA

```python
import pandas as pd 
import numpy as np

df_bcopa = pd.read_csv("../datasets/raw/balanced_copa.csv")
```

```python
df_bcopa['premise_length'] = df_bcopa['premise'].apply(lambda x: len(x))
df_bcopa['choice1_length'] = df_bcopa['choice1'].apply(lambda x: len(x))
df_bcopa['choice2_length'] = df_bcopa['choice2'].apply(lambda x: len(x))

first_filter_bcopa = (
    ((df_bcopa['premise_length'] > 20) & (df_bcopa['premise_length'] < 2000)) & 
    ((df_bcopa['choice1_length'] > 20) & (df_bcopa['choice1_length'] < 2000)) & 
    ((df_bcopa['choice2_length'] > 20) & (df_bcopa['choice2_length'] < 2000))
)
```

```python
df_bcopa = df_bcopa[first_filter_bcopa]
df_bcopa_cause = df_bcopa[df_bcopa['question'] ==  'cause']
df_bcopa_effect = df_bcopa[df_bcopa['question'] ==  'effect']
df_bcopa_cause_sampled = df_bcopa_cause.sample(n=200, random_state=42)
df_bcopa_effect_sampled = df_bcopa_effect.sample(n=200, random_state=42)
df_result_bcopa = pd.concat([df_bcopa_cause_sampled, df_bcopa_effect_sampled], ignore_index=True)
df_result_bcopa = df_result_bcopa.iloc[:, 0:7]
df_result_bcopa.head()
```

```python
df_result_bcopa.to_csv("../datasets/cleaned/cleaned_bcopa.csv", index=False)
```
