# Data Processing for PAWS

```python
import pandas as pd 
import numpy as np

df_paws = pd.read_csv("../datasets/paws.csv")
```

```python
df_paws['sentence1_length'] = df_paws['sentence1'].apply(lambda x: len(x))
df_paws['sentence2_length'] = df_paws['sentence2'].apply(lambda x: len(x))

# Apply the first filtering of Batayan
first_filter_paws = (
    ((df_paws['sentence1_length'] > 20) & (df_paws['sentence1_length'] < 2000)) & 
    ((df_paws['sentence2_length'] > 20) & (df_paws['sentence2_length'] < 2000))
)

df_paws = df_paws[first_filter_paws]
df_paws_true = df_paws[df_paws['label'] == 1]
df_paws_false = df_paws[df_paws['label'] == 0]

df_paws_true_sampled = df_paws_true.sample(n=200, random_state=42)
df_paws_false_sampled = df_paws_false.sample(n=200, random_state=42)
```

```python
df_result_paws = pd.concat([df_paws_true_sampled, df_paws_false_sampled], ignore_index=True)
print(len(df_result_paws))
df_result_paws = df_result_paws.iloc[:, 0:4]
df_result_paws.head()
```

```python
df_result_paws.to_csv("../datasets/cleaned/cleaned_paws.csv", index=False)
```
