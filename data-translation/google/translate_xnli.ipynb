```python
from dotenv import load_dotenv
import requests
import os
import pandas as pd

load_dotenv()
google_key = os.getenv("GOOGLE_KEY")
```

```python
df_xnli = pd.read_csv("../../datasets/cleaned/cleaned_xnli.csv")
```

```python
def google_translate(source_texts: list, key: str):
    url = "https://translation.googleapis.com/language/translate/v2"

    payload = {
        "q": source_texts,
        "target": "tl",
        "format": "text"
    }

    params = {"key": key}

    response = requests.post(url, params=params, json=payload)

    try:
        response_json = response.json()
        # print(response_json)
    except Exception:
        print("Non-JSON response:")
        print(response.text[:500])
        raise

    try:
        translations = [
            item["translatedText"]
            for item in response_json["data"]["translations"]
        ]
        return translations
    except Exception as e:
        print("Unexpected response structure:", e)
        print(response_json)
        raise
```

```python
sentence1 = df_xnli['sentence1'].to_list()
sentence2 = df_xnli['sentence2'].to_list()
```

```python
sentence1_translated_p1 = google_translate(sentence1[:100], google_key)
sentence1_translated_p2 = google_translate(sentence1[100:200], google_key)
sentence1_translated_p3 = google_translate(sentence1[200:300], google_key)
sentence1_translated_p4 = google_translate(sentence1[300:400], google_key)
sentence1_translated_p5 = google_translate(sentence1[400:500], google_key)
sentence1_translated_p6 = google_translate(sentence1[500:600], google_key)

sentence2_translated_p1 = google_translate(sentence2[:100], google_key)
sentence2_translated_p2 = google_translate(sentence2[100:200], google_key)
sentence2_translated_p3 = google_translate(sentence2[200:300], google_key)
sentence2_translated_p4 = google_translate(sentence2[300:400], google_key)
sentence2_translated_p5 = google_translate(sentence2[400:500], google_key)
sentence2_translated_p6 = google_translate(sentence2[500:600], google_key)
```

```python
result_sentence1 = pd.Series(sentence1_translated_p1 + sentence1_translated_p2 + sentence1_translated_p3 + sentence1_translated_p4 + sentence1_translated_p5 + sentence1_translated_p6)
result_sentence2 = pd.Series(sentence2_translated_p1 + sentence2_translated_p2 + sentence2_translated_p3 + sentence2_translated_p4 + sentence2_translated_p5 + sentence2_translated_p6)
df_xnli['sentence1'] = result_sentence1 
df_xnli['sentence2'] = result_sentence2
```

```python
df_xnli.to_csv('../../datasets/translated/google/google_translated_xnli.csv', index=False)
```
